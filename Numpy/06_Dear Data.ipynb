```python
import numpy as np
```

```python
# Data
sleep = np.array([4,5,1,2,6,4,3])
water = np.array([1.8,1.8,1.9,2.0,2.1,1.8,1.9])
yawns = np.array([14,14,12,20,32,12,32])
money = np.array([180, 200, 190, 250, 100, 90, 120])
```

```python
my_week = np.array([sleep, water, yawns, money])

WEEK = {
    'monday': 1,
    'tuesday': 2,
    'wednesday': 3,
    'thursday': 4,
    'friday': 5,
    'saturday': 6,
    'sunday': 7,
}

nama = input("Which day data you want to know? ").lower().strip()

if nama in WEEK:
    index = WEEK[nama] - 1
    labels = ['Sleep', 'Water', 'Yawns', 'Money']
    for label, row in zip(labels, my_week):
        print(f'{label} = {row[index]}')
else:
    print('day not found')
```
