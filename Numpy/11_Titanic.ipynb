## What is the average age of the passengers?

```python
ID = np.copy(passengers[:,0])
Survived = np.copy(passengers[:,1])
Class = np.copy(passengers[:,2])
Age = np.copy(passengers[:,3])

print('Average age of the passenger: ', np.mean(Age))
```

## What is the passenger number of the oldest passenger? Who is the youngest?

```python
oldest = np.max(Age)
youngest = np.min(Age)
data = dict(zip(Age, ID))
print('ID of the oldest passenger: ',data[oldest])
print('ID of the youngest passenger: ',data[youngest])
```

## What is the percentage of folks that survived?

```python
folks = 0
for i in Survived:
    if i == 1:
        folks += 1

print('Percentage of folks that survived: ', folks*2, '%')
```

## What is the percentage of the folks that survived based on their passenger class?

```python
upper, middle, lower = 0, 0, 0
for i in Class:
    if i == 1:
        upper += 1
    elif i == 2:
        middle += 1
    elif i == 3:
        lower += 1

print(f'Upper class: {upper*2}%')
print(f'Middle class: {middle*2}%')
print(f'Lower class: {lower*2}%')
```
